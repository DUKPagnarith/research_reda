# Section 3: Methodology

**Credit Risk and Dollarization in Cambodia: A Dual-Currency Analysis Using Interest Rate Spreads (2013–2025)**

---

This section presents the complete methodological framework for constructing currency-specific and system-wide Credit Risk Indices (CRIs) for Cambodia's dual-currency banking sector. The framework applies identically to both the USD and KHR segments, enabling direct cross-currency comparison.

## 3.1 Data Selection Rationale

The National Bank of Cambodia (NBC) publishes monthly interest rate data for deposit money banks (commercial and specialized banks), disaggregated by currency (KHR and USD), product type, and rate version. Three deliberate choices guide our data selection:

### Rate Version: New Amount over Outstanding Amount

The NBC reports two versions of each interest rate:

| Version | What It Measures | Speed of Response |
|---------|-----------------|-------------------|
| **Weighted Average Rate on New Amount** | Rates on newly issued loans/deposits in that month | Fast — reflects banks' current risk pricing |
| Weighted Average Rate on Outstanding Amount | Average rate across all existing loans/deposits | Slow — diluted by legacy rates from prior years |

We use **New Amount rates** for primary analysis because an effective early warning indicator requires the series that reacts fastest to changing risk conditions. When a bank perceives rising credit risk, it immediately adjusts rates on new loans; outstanding amount rates remain anchored by legacy contracts issued months or years earlier. Outstanding Amount rates are retained for robustness checks (Section 3.12).

### Product Selection: Term Loans and Term Deposits

**Loan products:**

| Product | Selected? | Rationale |
|---------|:---------:|-----------|
| **Term Loans** | ✅ | Most representative of standard credit risk pricing; fixed maturity with pricing directly reflecting credit assessment |
| Overdraft | ✗ | Priced for flexibility and liquidity provision, not pure credit risk |
| Credit Card | ✗ | Bundles convenience premiums and unsecured risk; pricing distorts the credit risk signal |
| Other Loans | ✗ | Ambiguous composition makes interpretation unreliable |

**Deposit products:**

| Product | Selected? | Rationale |
|---------|:---------:|-----------|
| **Term Deposits** | ✅ | Best maturity match with term loans; pricing reflects time value of money |
| Demand Deposits | ✗ | Near-zero rates driven by liquidity needs, not credit dynamics |
| Saving Deposits | ✗ | Partially liquidity-driven; poor maturity match |
| Other Deposits | ✗ | Ambiguous composition |

Using term loans and term deposits for both currencies ensures a clean, consistent spread that isolates credit risk compensation as closely as possible.

## 3.2 Interest Rate Spread Definition

For each currency $c \in \{\text{USD}, \text{KHR}\}$, we compute the credit spread as:

$$
S_t^c = r_{\text{Term Loans},t}^{c} - r_{\text{Term Deposits},t}^{c}
$$

where:
- $r_{\text{Term Loans},t}^{c}$ = weighted average rate on new term loans in currency $c$ at month $t$
- $r_{\text{Term Deposits},t}^{c}$ = weighted average rate on new term deposits in currency $c$ at month $t$

This yields two time series: $S_t^{\text{USD}}$ and $S_t^{\text{KHR}}$, each comprising **156 monthly observations** (January 2013 – December 2025).

### Why Model Each Currency Separately?

USD and KHR lending operate under fundamentally different risk environments:

- **USD lending** is influenced by global dollar conditions, U.S. Federal Reserve monetary policy, and international capital flows
- **KHR lending** is more sensitive to domestic monetary policy, NBC reserve requirement changes, and exchange rate expectations

Just as credit risk practitioners build separate models for different portfolio segments (a principle established in the Basel framework), we construct separate CRIs for each currency to capture these distinct dynamics.

## 3.3 The Ornstein–Uhlenbeck (OU) Stochastic Model

For each currency $c$, we model the spread as a **mean-reverting diffusion process** governed by the stochastic differential equation (SDE):

$$
dS_t^c = \kappa^c\left(\theta^c - S_t^c\right)dt + \sigma^c\,dW_t^c
$$

where $W_t^c$ is a standard Wiener process (Brownian motion).

### Parameter Interpretation

| Parameter | Symbol | Economic Meaning | Cross-Currency Comparison |
|-----------|:------:|-----------------|-------------------------|
| Long-run equilibrium | $\theta^c$ | The "normal" spread level to which the process gravitates | Is the structural spread higher for KHR or USD? |
| Mean-reversion speed | $\kappa^c$ | How quickly the spread returns to equilibrium after a shock | Which currency recovers faster from credit stress? |
| Volatility | $\sigma^c$ | The magnitude of random fluctuations around the mean | Which currency exhibits more spread instability? |

### Why the OU Process?

1. **Mean reversion is economically natural**: Interest rate spreads do not wander randomly — competitive forces, monetary policy, and arbitrage pull them back toward a long-run level. The OU process was first applied to finance by Vasicek (1977) precisely for this reason.

2. **Analytical tractability**: The OU process has a **closed-form transition density** (Gaussian), enabling exact Maximum Likelihood Estimation and analytical computation of crisis probabilities without Monte Carlo simulation.

3. **Parsimony**: Only three parameters per currency, estimated from 156 observations — a well-identified estimation problem.

### Explicit Solution

The SDE has the closed-form solution:

$$
S_t^c = \theta^c + \left(S_0^c - \theta^c\right)e^{-\kappa^c t} + \sigma^c\int_0^t e^{-\kappa^c(t-s)}\,dW_s^c
$$

Since the stochastic integral of a deterministic function with respect to Brownian motion is Gaussian, the spread at time $t$ follows a **normal distribution**:

$$
S_t^c \sim \mathcal{N}\!\left(m^c(t),\; v^c(t)\right)
$$

with conditional mean and variance:

$$
m^c(t) = \theta^c + \left(S_0^c - \theta^c\right)e^{-\kappa^c t}
$$

$$
v^c(t) = \frac{(\sigma^c)^2}{2\kappa^c}\left(1 - e^{-2\kappa^c t}\right)
$$

### Half-Life of Mean Reversion

A useful derived quantity is the **half-life** — the time for a shock to decay to half its initial magnitude:

$$
t_{1/2}^c = \frac{\ln 2}{\kappa^c}
$$

This provides an intuitive measure: a half-life of 6 months means that if the spread is shocked to 2 percentage points above equilibrium, it will be approximately 1 percentage point above equilibrium six months later.

## 3.4 Parameter Estimation: Maximum Likelihood Estimation (MLE)

### Transition Density

For discretely observed data with equal time spacing $\Delta t = 1/12$ (one month), the **exact transition density** of the OU process gives:

$$
S_{t_{i+1}}^c \mid S_{t_i}^c \sim \mathcal{N}\!\left(\mu_i^c,\;\gamma^{c\,2}\right)
$$

where:

$$
\mu_i^c = \theta^c + \left(S_{t_i}^c - \theta^c\right)e^{-\kappa^c \Delta t}
$$

$$
\gamma^{c\,2} = \frac{(\sigma^c)^2}{2\kappa^c}\left(1 - e^{-2\kappa^c \Delta t}\right)
$$

### Log-Likelihood Function

Given $n$ transitions $(S_{t_0}, S_{t_1}, \ldots, S_{t_n})$, the log-likelihood is:

$$
\ell\!\left(\kappa^c, \theta^c, \sigma^c\right) = -\frac{n}{2}\ln(2\pi) - \frac{n}{2}\ln\!\left(\gamma^{c\,2}\right) - \sum_{i=1}^{n}\frac{\left(S_{t_i}^c - \mu_{i-1}^c\right)^2}{2\,\gamma^{c\,2}}
$$

We maximize this function using numerical optimization (`scipy.optimize.minimize` with the Nelder-Mead or L-BFGS-B method applied to the **negative** log-likelihood) to obtain the MLE estimates $\hat{\kappa}^c$, $\hat{\theta}^c$, and $\hat{\sigma}^c$ for each currency.

### Confidence Intervals

Standard errors are computed from the inverse of the observed Fisher information matrix (the Hessian of the negative log-likelihood at the optimum). Approximate 95% confidence intervals are constructed as $\hat{\psi} \pm 1.96 \cdot \text{SE}(\hat{\psi})$ for each parameter $\psi \in \{\kappa, \theta, \sigma\}$.

### Cross-Check: AR(1) Regression

The OU process observed at discrete intervals is equivalent to an **AR(1) model**:

$$
S_{t+1}^c = a^c + b^c S_t^c + \varepsilon_t^c, \qquad \varepsilon_t^c \sim \mathcal{N}(0, \gamma^{c\,2})
$$

with the mapping:

| AR(1) Parameter | OU Equivalent |
|:---------------:|:-------------:|
| $b^c = e^{-\kappa^c \Delta t}$ | $\kappa^c = -\frac{\ln b^c}{\Delta t}$ |
| $a^c = \theta^c(1 - b^c)$ | $\theta^c = \frac{a^c}{1 - b^c}$ |
| $\text{Var}(\varepsilon^c) = \frac{(\sigma^c)^2}{2\kappa^c}(1 - (b^c)^2)$ | $\sigma^c = \sqrt{\frac{2\kappa^c \cdot \text{Var}(\varepsilon^c)}{1 - (b^c)^2}}$ |

A simple OLS regression of $S_{t+1}$ on $S_t$ provides an independent validation of the MLE estimates. Agreement between the two methods adds credibility to the parameter estimates.

## 3.5 Crisis Threshold Definition

For each currency, we define a **critical threshold** $S_c^c$ above which the spread signals crisis-level credit risk conditions:

$$
S_c^{\text{USD}} = P_{95}\!\left(S_t^{\text{USD}}\right), \qquad S_c^{\text{KHR}} = P_{95}\!\left(S_t^{\text{KHR}}\right)
$$

where $P_{95}(\cdot)$ denotes the 95th percentile of the historical spread distribution.

**Currency-specific thresholds are essential**: because KHR spreads are structurally wider than USD spreads (reflecting the exchange rate risk premium embedded in riel lending), a single absolute threshold would be economically meaningless. Each currency is measured against its own historical distribution.

**Sensitivity analysis**: We also compute thresholds at the 90th and 99th percentiles to verify that the CRI's qualitative behavior is robust to threshold choice.

## 3.6 Crisis Probability

Using the OU model's Gaussian transition density, we compute the **probability that the spread exceeds the crisis threshold** at each point in time:

$$
P\!\left(S_t^c > S_c^c\right) = 1 - \Phi\!\left(\frac{S_c^c - m^c(t)}{\sqrt{v^c(t)}}\right)
$$

where:
- $\Phi(\cdot)$ is the cumulative distribution function (CDF) of the standard normal distribution
- $m^c(t)$ is the conditional mean of the spread at time $t$
- $v^c(t)$ is the conditional variance of the spread at time $t$

This provides a **time-varying, model-implied probability** of entering credit stress conditions for each currency. The analytical nature of this computation (no simulation required) is a key advantage of the OU framework.

## 3.7 Currency-Specific Credit Risk Index (CRI)

For each currency $c$, the Credit Risk Index combines normalized volatility with crisis probability:

$$
\text{CRI}^c = \alpha \cdot \hat{\sigma}^c_{\text{norm}} + \beta \cdot P\!\left(S_t^c > S_c^c\right)
$$

where:
- $\hat{\sigma}^c_{\text{norm}}$ is the normalized OU volatility parameter (capturing baseline spread instability)
- $P(S_t^c > S_c^c)$ is the model-implied crisis probability (capturing proximity to crisis conditions)

**Baseline weights**: $\alpha = \beta = 0.5$ (equal weight to volatility and crisis probability).

The CRI captures two complementary dimensions of risk:
1. **Volatility component** ($\hat{\sigma}^c$): measures how erratically spreads move — high volatility signals an unstable risk environment
2. **Crisis probability component**: measures how close the current spread is to historically extreme levels — high probability signals imminent stress

**Sensitivity**: Alternative weight configurations ($\alpha = 0.3, \beta = 0.7$ and $\alpha = 0.7, \beta = 0.3$) are tested in the robustness analysis.

## 3.8 System-Wide Composite CRI

The individual currency CRIs are aggregated into a single system-level index:

$$
\text{CRI}_{\text{System}} = w_{\text{USD}} \cdot \text{CRI}^{\text{USD}} + w_{\text{KHR}} \cdot \text{CRI}^{\text{KHR}}
$$

Three weighting schemes are considered:

| Weighting Approach | $w_{\text{USD}}$ | $w_{\text{KHR}}$ | Rationale |
|:------------------:|:----------------:|:----------------:|-----------|
| **Loan-share weighted** (primary) | ~0.80 | ~0.20 | Reflects the actual composition of Cambodia's banking system |
| Equal weighted | 0.50 | 0.50 | Gives equal attention to both currency segments |
| Time-varying | Changes annually | Changes annually | Captures the gradual de-dollarization trend |

The **loan-share weighted** scheme is our primary specification because it reflects the economic reality: a shock to USD lending (80% of the system) has a far larger systemic impact than an equivalent shock to KHR lending (20%). The equal-weighted and time-varying alternatives serve as robustness checks.

## 3.9 Stress Testing

We apply multiplicative stress shocks to the observed spreads to simulate adverse credit conditions:

$$
S_t^{c,\text{stress}} = S_t^c\left(1 + \delta\right)
$$

Three stress scenarios are considered:

| Scenario | Shock ($\delta$) | Economic Interpretation |
|:--------:|:----------------:|------------------------|
| Mild | 0.10 (10%) | Minor economic slowdown or sectoral stress |
| Moderate | 0.30 (30%) | Significant financial stress (comparable to a regional crisis) |
| Severe | 0.50 (50%) | Major crisis scenario (comparable to the GFC or severe pandemic impact) |

Under each scenario, we recompute $\text{CRI}^{\text{USD}}$, $\text{CRI}^{\text{KHR}}$, and $\text{CRI}_{\text{System}}$ and compare them to the baseline (unstressed) values.

**Key policy question**: Is the USD segment or KHR segment more fragile under stress? If the dominant USD segment (80% of lending) is more stress-sensitive, the current dollarized structure poses greater systemic risk. If the KHR segment is more sensitive, this informs the risk management of the de-dollarization transition.

## 3.10 COVID-19 Sub-Period Analysis

The 13-year sample (2013–2025) spans a major exogenous shock — the COVID-19 pandemic — which serves as a **natural experiment** to test the CRI framework's responsiveness. We divide the sample into three sub-periods:

| Period | Dates | Label | Observations |
|:------:|:-----:|:-----:|:------------:|
| Pre-COVID | Jan 2013 – Dec 2019 | Baseline | 84 |
| COVID | Jan 2020 – Dec 2021 | Shock | 24 |
| Post-COVID | Jan 2022 – Dec 2025 | Recovery | 48 |

For each sub-period and each currency, we:
1. Re-estimate the OU parameters ($\kappa^c, \theta^c, \sigma^c$) via MLE
2. Recompute crisis probabilities and CRIs
3. Compare parameter shifts across periods

**Key questions**:
- Did the COVID-19 shock cause a **temporary** or **structural** shift in credit risk parameters?
- Which currency segment was more affected by the pandemic?
- Have the risk dynamics returned to pre-COVID levels, or has a new equilibrium emerged?

The 84-observation pre-COVID baseline provides a strong reference for identifying pandemic-induced changes.

## 3.11 Rolling Window Analysis

To track the **time evolution** of credit risk dynamics, we implement a **36-month rolling window** re-estimation:

- At each month $t$ (starting from month 37), we estimate the OU parameters using only the most recent 36 observations: $\{S_{t-35}^c, S_{t-34}^c, \ldots, S_t^c\}$
- This produces time series of parameter estimates: $\hat{\kappa}_t^c$, $\hat{\theta}_t^c$, $\hat{\sigma}_t^c$

### Why 36 Months?

The choice of a 36-month (3-year) window balances two competing objectives:

| Consideration | Shorter Window (e.g., 24 months) | 36-Month Window |
|--------------|----------------------------------|------------------|
| Responsiveness | More responsive to recent changes | Moderately responsive |
| Estimation stability | Noisy — only 23 transitions for 3 parameters; log-likelihood surface becomes flat | 35 transitions provide statistically stable MLE estimates |
| Coverage | Output starts later, more history lost | Good balance of history and output length |

With 36 observations (35 transitions), the MLE estimation for 3 parameters is well-identified and produces reliable estimates.

### Output

The rolling window analysis produces time-varying plots showing:
- Whether the banking sector is becoming more or less stable over time
- Whether USD and KHR risk dynamics are converging or diverging
- Whether specific events (COVID, Fed tightening, NBC policy changes) cause visible parameter shifts

## 3.12 Robustness Checks

We perform a battery of robustness checks to verify that the main findings are not artifacts of specific methodological choices:

| Check | What Is Varied | Purpose |
|-------|:--------------:|---------|
| **Alternative rate measure** | Outstanding Amount rates instead of New Amount | Confirms results hold regardless of rate version |
| **Alternative crisis thresholds** | 90th and 99th percentiles instead of 95th | Shows CRI behavior is not threshold-sensitive |
| **Alternative CRI component weights** | ($\alpha=0.3, \beta=0.7$) and ($\alpha=0.7, \beta=0.3$) | Shows the relative ranking of USD vs. KHR is stable |
| **Alternative system weights** | Equal (0.50/0.50) vs. loan-share (0.80/0.20) | Shows system CRI is robust to aggregation method |

If the principal findings — the relative risk ranking of USD and KHR, the temporal pattern of CRI, and the directional response to COVID-19 — are preserved across these variations, the framework's validity is reinforced.

# Extended Analyses

## 3.13 Regime-Switching Ornstein–Uhlenbeck Model

The standard OU model assumes constant parameters throughout the sample. To allow for **structural breaks** in credit risk dynamics, we extend the framework to a **two-state Markov Regime-Switching OU (RS-OU) model**:

$$
dS_t^c = \kappa_{s_t}^c\!\left(\theta_{s_t}^c - S_t^c\right)dt + \sigma_{s_t}^c\,dW_t^c, \qquad s_t \in \{1, 2\}
$$

where the latent state $s_t$ follows a first-order Markov chain with transition probability matrix:

$$
\mathbf{P} = \begin{pmatrix} p_{11} & 1 - p_{11} \\ 1 - p_{22} & p_{22} \end{pmatrix}
$$

- **State 1** ("Calm"): lower volatility, stable spread dynamics
- **State 2** ("Stress"): higher volatility, potentially different mean-reversion behavior

### Estimation

Parameters are estimated using the **Hamilton filter** (forward pass) combined with the **Kim smoother** (backward pass) to obtain smoothed state probabilities. Model selection between the single-regime OU and the RS-OU is based on the **Bayesian Information Criterion (BIC)**.

### Value Added

The RS-OU model answers: *Do USD and KHR credit risk operate under the same regime structure?* If the KHR segment switches between calm and stress regimes more frequently, it corroborates the finding of higher KHR instability from the standard OU analysis.

## 3.14 Dynamic Threshold CRI

The baseline CRI uses a **static** crisis threshold (the 95th percentile of the full sample). An alternative is to use a **rolling 36-month threshold** that adapts to the evolving distribution of spreads:

$$
S_{c,t}^c = P_{95}\!\left(\{S_{t-35}^c, \ldots, S_t^c\}\right)
$$

This dynamic threshold:
- Accounts for the **structural compression** of KHR spreads over time (a static threshold based on early high-spread observations may be irrelevant for the post-2020 environment)
- Provides a more **locally calibrated** measure of what constitutes "abnormal" spread behavior
- Is compared directly against the static-threshold CRI to assess whether the temporal pattern of risk changes materially

## 3.15 Implied Default Probability

We extract **market-implied default probabilities** from the credit spreads using a reduced-form credit risk framework. Under the assumption that the spread compensates for expected losses:

$$
S_t^c \approx h_t^c \cdot (1 - R^c)
$$

where:
- $h_t^c$ is the instantaneous hazard rate (default intensity) in currency $c$
- $R^c$ is the recovery rate (assumed $R = 0.40$ as the standard Basel assumption)

The **hazard rate** is extracted as:

$$
h_t^c = \frac{S_t^c}{1 - R^c}
$$

The **1-year cumulative default probability** is then:

$$
\text{PD}_{1\text{yr}}^c(t) = 1 - \exp\!\left(-h_t^c\right)
$$

### Recovery Rate Sensitivity

Since the recovery rate is assumed rather than observed, we perform a sensitivity analysis across $R \in \{0.20, 0.30, 0.40, 0.50, 0.60\}$ to show how implied PD estimates vary with this assumption. This is particularly important in the Cambodian context where empirical recovery rate data is scarce.

## 3.16 Policy Divergence Analysis (VAR Framework)

To examine the **macroeconomic transmission channels** affecting credit risk in Cambodia's dual-currency system, we employ a **Vector Autoregression (VAR)** model:

$$
\mathbf{Y}_t = \mathbf{c} + \sum_{j=1}^{p} \mathbf{A}_j \mathbf{Y}_{t-j} + \mathbf{u}_t
$$

where the variable vector is:

$$
\mathbf{Y}_t = \begin{pmatrix} S_t^{\text{USD}} \\ S_t^{\text{KHR}} \\ \text{FedFunds}_t \end{pmatrix}
$$

and $p$ is the optimal lag order selected by the Bayesian Information Criterion (BIC).

### Analytical Tools

1. **Granger Causality Tests**: Does the Fed Funds Rate Granger-cause movements in USD or KHR spreads? Does one currency's spread predict the other's?

2. **Impulse Response Functions (IRFs)**: How do USD and KHR credit spreads respond dynamically to a one-standard-deviation shock in the Fed Funds Rate? How long does the effect persist?

3. **Variance Decomposition**: What proportion of the forecast error variance in each spread is attributable to Fed policy shocks versus innovations in the other currency's spread?

### Policy Relevance

This analysis directly tests whether Cambodia has effectively **outsourced its monetary policy** to the U.S. Federal Reserve through dollarization. If USD spreads respond strongly to Fed shocks while KHR spreads do not, it provides empirical support for the argument that de-dollarization would enhance NBC's monetary policy autonomy.
